```python
import os
import json
import random
import time

# Function to write data to a JSON file
def write_json(data, file_path):
    with open(file_path, 'a') as file:
        for item in data:
            json.dump(item, file)
            file.write('\n')

# Function to process and split data into train and test sets
def process_chunk(chunk, train_file, test_file, train_percentage=0.8):
    train_data = []
    test_data = []
    for line in chunk:
        try:
            json_data = json.loads(line)
            if random.random() < train_percentage:
                train_data.append(json_data)
            else:
                test_data.append(json_data)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")

    return train_data, test_data

# Function to get the size of a file
def get_file_size(file_path):
    return os.path.getsize(file_path)

# Path to the directory containing JSON files
directory = '/Users/kravisankaran/Desktop/data'

# Record the start time
start_time = time.time()

# List to store chunks from each file
all_chunks = []

# Process each JSON file
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        file_path = os.path.join(directory, filename)
        
        # Get the size of the file
        file_size = get_file_size(file_path)
        chunk_size = min(100 * 1024 * 1024, file_size)  # Chunk size of 100 MB or the file size, whichever is smaller
        
        print(f"Processing data from file: {filename}")
        
        with open(file_path, 'r') as file:
            chunk = ''
            for line in file:
                chunk += line
                if len(chunk) >= chunk_size:
                    all_chunks.append((filename, chunk))
                    chunk = ''
            if chunk:
                all_chunks.append((filename, chunk))

# Shuffle the list of chunks
random.shuffle(all_chunks)

# Process shuffled chunks
train_file = "combined_train_data_chunked3.json"
test_file = "combined_test_data_chunked3.json"

for filename, chunk in all_chunks:
    train_data, test_data = process_chunk(chunk.split('\n'), train_file, test_file)
    write_json(train_data, train_file)
    write_json(test_data, test_file)

# Record the time after processing
end_time = time.time()

print("Data processing completed.")

# Print the time taken
print(f"Time taken: {end_time - start_time} seconds")
```
