```python
import os
import json
import random

# Function to read data from JSON files
def read_json(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

# Function to write data to a JSON file
def write_json(data, file_path):
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

# List to store combined data
combined_data = []

# Path to the directory containing JSON files
directory = '/Users/kravisankaran/Desktop/data'

# Iterate over each JSON file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        file_path = os.path.join(directory, filename)
        
        # Read JSON content as lines
        with open(file_path, 'r') as file:
            lines = file.readlines()
        
        # Iterate over each line and load JSON data
        for line in lines:
            try:
                data = json.loads(line)
                combined_data.append(data)
            except json.JSONDecodeError as e:
                print(f"Error reading file {filename}: {e}")

# Shuffle combined data
random.shuffle(combined_data)

# Write combined data to a new JSON file
write_json(combined_data, "combined_data3.json")

print("Combined data written to combined_data3.json")

```

```python
import os
import json
import random

# Function to read data from JSON files
def read_json(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            try:
                json_data = json.loads(line)
                data.append(json_data)
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON in file {file_path}: {e}")
    return data

# Function to write data to a JSON file
def write_json(data, file_path):
    with open(file_path, 'w') as file:
        for item in data:
            json.dump(item, file)
            file.write('\n')

# List to store combined data
combined_data = []

# Path to the directory containing JSON files
directory = '/Users/kravisankaran/Desktop/data'

# Iterate over each JSON file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        file_path = os.path.join(directory, filename)
        data = read_json(file_path)
        
        # Shuffle the data randomly
        random.shuffle(data)
        
        # Calculate 80% of records
        split_index = int(0.8 * len(data))
        
        # Select 80% of records randomly
        selected_data = data[:split_index]
        
        # Add selected data to combined_data list
        combined_data.extend(selected_data)

# Shuffle combined data
random.shuffle(combined_data)

# Write combined data to a new JSON file
write_json(combined_data, "combined_data6.json")

print("Combined data written to combined_data6.json")
```
