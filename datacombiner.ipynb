```python
# Without chunking - and excluding Kindle and PetSupplies and only combining 
# Processing data from file: Toys_and_Games_5.json
# Processing data from file: AMAZON_FASHION_5.json
# Processing data from file: Cell_Phones_and_Accessories_5.json
# Processing data from file: Appliances_5.json
# Processing data from file: Arts_Crafts_and_Sewing_5.json
# Processing data from file: Automotive_5.json
# Processing data from file: Grocery_and_Gourmet_Food_5.json
# Processing data from file: CDs_and_Vinyl_5.json
# Processing data from file: Tools_and_Home_Improvement_5.json
# Processing data from file: All_Beauty_5.json
import os
import json
import random
import time

# Function to write data to a JSON file
def write_json(data, file_path):
    with open(file_path, 'w') as file:
        for item in data:
            json.dump(item, file)
            file.write('\n')

# List to store combined training and test data
combined_train_data = []
combined_test_data = []

# Path to the directory containing JSON files
directory = '/Users/kravisankaran/Desktop/data'

# Record the start time
start_time = time.time()

# Iterate over each JSON file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        file_path = os.path.join(directory, filename)
        print(f"Processing data from file: {filename}")
        
        # List to store data from current file
        file_data = []
        
        # Open file for reading
        with open(file_path, 'r') as file:
            # Iterate over each line in the file
            for line in file:
                try:
                    json_data = json.loads(line)
                    file_data.append(json_data)
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON in file {file_path}: {e}")
        
        # Calculate 80% of records from current file for training set
        split_index = int(0.8 * len(file_data))
        
        # Select 80% of records randomly from current file for training set
        train_data = file_data[:split_index]
        
        # Select rest 20% of records from current file for test set
        test_data = file_data[split_index:]
        
        # Add selected data to combined training and test data
        combined_train_data.extend(train_data)
        combined_test_data.extend(test_data)

# Shuffle combined training and test data
random.shuffle(combined_train_data)
random.shuffle(combined_test_data)

# Write combined training data to a new JSON file
write_json(combined_train_data, "combined_train_data.json")

# Write combined test data to a new JSON file
write_json(combined_test_data, "combined_test_data.json")

# Record the time after writing to files
write_time = time.time()

print("Combined training data written to combined_train_data.json")
print("Combined test data written to combined_test_data.json")

# Print the time taken
print(f"Time taken: {write_time - start_time} seconds")
```
